# Example

## Let's import [Pokemon Images](https://www.kaggle.com/datasets/kvpratama/pokemon-images-dataset) dataset from Kaggle:

```python
! kaggle datasets download kvpratama/pokemon-images-dataset
```

```python
! mkdir pokemon_images
```

```python
! unzip -q pokemon-images-dataset.zip -d pokemon_images
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('/content/pokemon_images/pokemon/pokemon/25.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread('/content/pokemon_images/pokemon_jpg/pokemon_jpg/25.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```
